```python
#DFS(深さ優先探索) アルゴリズムによる8パズルの解法

'''
8パズルに関しては https://github.com/NiloofarShahbaz/8-puzzle-search-implementation
FIFO（fisrt-in-first-out）はスタックを採用し、クラスの作成はこれを参考した https://yu-nix.com/archives/python-my-puts/
これを参考に一部改変した

'''

#ここから転載
from time import time
class Puzzle:
    goal_state=[1,0,3,4,2,6,7,5,8]#ゴールとする状態を定義
    heuristic=None
    evaluation_function=None
    needs_hueristic=False
    num_of_instances=0
    def __init__(self,state,parent,action,path_cost,needs_hueristic=False):
        self.parent=parent
        self.state=state
        self.action=action
        if parent:
            self.path_cost = parent.path_cost + path_cost
        else:
            self.path_cost = path_cost
        
        Puzzle.num_of_instances+=1

    def __str__(self):
        return str(self.state[0:3])+'\n'+str(self.state[3:6])+'\n'+str(self.state[6:9])

    def goal_test(self):
        if self.state == self.goal_state:
            return True
        return False

    @staticmethod
    def find_legal_actions(i,j):
        legal_action = ['U', 'D', 'L', 'R']
        if i == 0:  # up is disable
            legal_action.remove('U')
        elif i == 2:  # down is disable
            legal_action.remove('D')
        if j == 0:
            legal_action.remove('L')
        elif j == 2:
            legal_action.remove('R')
        return legal_action

    def generate_child(self):#子節点の生成
        children=[]
        x = self.state.index(0)
        i = int(x / 3)
        j = int(x % 3)
        legal_actions=self.find_legal_actions(i,j)

        for action in legal_actions:
            new_state = self.state.copy()
            if action == 'U':#0を上に移動
                new_state[x], new_state[x-3] = new_state[x-3], new_state[x]
            elif action == 'D':#0を下に移動
                new_state[x], new_state[x+3] = new_state[x+3], new_state[x]
            elif action == 'L':#0を左に移動
                new_state[x], new_state[x-1] = new_state[x-1], new_state[x]
            elif action == 'R':#0を右に移動
                new_state[x], new_state[x+1] = new_state[x+1], new_state[x]
            children.append(Puzzle(new_state,self,action,1,self.needs_hueristic))
        return children

    def find_solution(self):#状態が解かどうかを調べ、親を辿っていき上下左右を確認する
        solution = []
        solution.append(self.action)
        path = self
        while path.parent != None:
            path = path.parent
            solution.append(path.action)
        solution = solution[:-1]
        solution.reverse()
        return solution
#ここまで転載

#ここからスタッフのクラス（自作）

class MyStack:#スタックに関するクラス
    def __init__(self):
        self.stack = []#スタック（箱）
    def push(self,item):
        self.stack.append(item)#アイテムの追加
    def pop(self):#アイテムの取り出し
        if len(self.stack) == 0:
            return None
        return self.stack.pop()
    def isEmpty(self):#スタックの中が空かどうかを調べる
        if len(self.stack) == 0:
            return True
        else:
            return False
            
#自作関数  
def depth_first_search(initial_state):
    start_node = Puzzle(initial_state, None, None,0)
    if start_node.goal_test():
        return start_node.find_solution()
    stack = MyStack()
    stack.push(start_node)#キューに最初のノードを追加
    explored=[]#探索済みの要素を格納する配列
    while not(stack.isEmpty()): #キューが空でないとき
        node=stack.pop()#キューからアイテムを取り出す
        
        if not(node in explored):#まだ未探索のキューか確認する
            explored.append(node)#探索済みとして格納
            if node.goal_test():
                return node.find_solution()#そのキューが答えのときwhile を抜ける
                break
            
        #もし答えじゃないときは、子を生成する
            children=node.generate_child()#探索済みに格納した状態から、4（3）つの子状態を作成する
            for child in reversed(children): #未探索の配列に子要素を格納
                if not (child in explored):#新しい生成した子が、探検済みかどうかを判定
                    stack.push(child)
                    
    return



state=[[1, 2, 3,
        4, 5, 6,
        7, 0, 8]]

for i in state:
    Puzzle.num_of_instances=0
    t0=time()
    dfs=depth_first_search(i)
    t1=time()-t0
    print('DFS:', dfs)
    print('space:',Puzzle.num_of_instances)
    print('time:',t1)
    print()
    print('------------------------------------------')
```
